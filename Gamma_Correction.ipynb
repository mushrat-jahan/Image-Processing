```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
input_image = cv2.imread('/content/drive/MyDrive/Demo/2.png')
# Define the gamma correction factor
gamma = 1.5  # You can adjust this value to control the correction
# Perform gamma correction
gamma_corrected = np.power(input_image / 255.0, gamma) * 255.0
# Convert the gamma corrected image to 8-bit unsigned integer format
gamma_corrected = np.uint8(gamma_corrected)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected)
plt.title('Gamma Correction Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

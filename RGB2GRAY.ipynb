```python
import cv2
import os
import glob
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img_path = r'/content/drive/MyDrive/Dataset-Resize/demo/kaggle-resize/3'
des_path = r'/content/drive/MyDrive/GAN/demo/kaggle.gray/4'
```

```python
i=1
for img in glob.glob(img_path+'/*.*'):
    try:
        image =cv2.imread(img)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        os.chdir(des_path)
        filename = 'gray_0Normal_'+str(i)+'.png'
        cv2.imwrite(filename, gray)
        i+=1
    except:
        print('Operation not executed')
```

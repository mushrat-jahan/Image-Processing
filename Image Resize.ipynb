```python
from PIL import Image
import glob
```

```python
image_list = []
resized_images = []
for filename in glob.glob('/content/drive/MyDrive/Knee work/crop/4/*.png'):
  print(filename)
  img = Image.open(filename)
  image_list.append(img)

for image in image_list:
  image.show()
  image = image.resize((224,224))
  resized_images.append(image)
for (i,new) in enumerate(resized_images):
    new.save('{}{}{}'.format("/content/drive/MyDrive/Knee work/Resize/4/",i+1,'.png'))
```

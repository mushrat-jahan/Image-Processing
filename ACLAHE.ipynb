```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
# Load the image in grayscale
input_image = cv2.imread('/content/drive/MyDrive/Demo/2.png', cv2.IMREAD_GRAYSCALE)
# Apply Adaptive Contrast-Limited Adaptive Histogram Equalization (ACLAHE)
ahe_image = exposure.equalize_adapthist(input_image, clip_limit=0.03)
# Display the original and ACLAHE images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(ahe_image, cmap='gray')
plt.title('ACLAHE Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

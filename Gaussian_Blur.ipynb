```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
# Load an image from file
image = cv2.imread('/content/drive/MyDrive/Github/1.png')
# Define the kernel size for the Gaussian blur (should be an odd number)
kernel_size = (5, 5)
# Apply Gaussian blur to the image
blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
# Save the blurred image to a file
cv2.imwrite('output_blurred.png', blurred_image)
# Display the original and blurred images (optional)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Gaussian Blur Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```
